### **Prediction of Song Genre Based on Audio Features**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from sklearn import tree
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn.metrics import accuracy_score,classification_report

from IPython.display import Image  
import pydotplus

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
```

```python
os.getcwd()
```

```python
dataUnstructured = pd.read_csv("data\spotify.csv") 
dataUnstructured = dataUnstructured.iloc[:,1:] #The first column was just row indices, so I removed it.
```

```python
for i in range(0,len(dataUnstructured.columns)):
    missing = [0] * len(dataUnstructured.columns)
    missing[i] = dataUnstructured.iloc[:,i].isna().sum()

pd.DataFrame({"Column":dataUnstructured.columns,"# of Missing Rows":missing})
```

```python
relevant_columns=["genre","acousticness","danceability","energy","instrumentalness",
                   "liveness","loudness","speechiness","tempo","valence"]
data=dataUnstructured.loc[:,relevant_columns]
data=data[data["genre"] != "Gaming"] 
```

The following table shows the mean of the audio features by genre:

```python
data.groupby("genre").agg("mean").round(3).reset_index()
```

In terms of audio features, K-Pop and Pop are very similar across the board. They will be treated as one genre.

```python
data["genre"].replace({"K-pop":"Pop","reggae":"Reggae"},inplace=True)
```

```python
dummy=np.array([
    ["acousticness","danceability","energy"],
    ["instrumentalness","liveness","valence"],
    ["speechiness","loudness","tempo"]
    ])

fig,axis=plt.subplots(3,3,figsize=(10,10)) 
fig.subplots_adjust(hspace=0.65,wspace=0.65)

for i in range(0,3):
    for j in range(0,3):
        axis[i,j].set_title(dummy[i,j])

for i in range(0,3):
    for j in range(0,3):
        axis[i,j].hist(data[dummy[i,j]])
```

#### **Model 1: Classification Tree**

```python
y=data["genre"]
X=data.iloc[:,1:]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

First, a decision tree is created without varying any parameters:

```python
classifier=DecisionTreeClassifier(random_state=42)
default_model=classifier.fit(X_train,y_train)
default_model
```

```python
predictions=default_model.predict(X_test)
print(classification_report(y_test,predictions))
```

- **Support** is simply the observations that belong to a particular genre.
- **Accuracy** is the proportion of correct predictions out of the entire test set. In this case, the accuracy of the model is $58\%$.
- For each song genre, **precision is the proportion of true positives out of the sum of true positives and false positives**. Consider classical music with a precision of $0.81$: $81\%$ of all songs predicted to be classical music were correct,

$$ 
\frac{\text{songs correctly predicted as classical}}{\text{songs correctly predicted as classical + songs incorrectly predicted as classical}}
$$

- For each song genre, **recall is the proportion of true positives out of the sum of true positives and false negatives**. Consider classical music with a recall of $0.79$: $79\%$ of all classical music were correctly predicted as such,

$$ 
\frac{\text{songs correctly predicted as classical}}{\text{songs correctly predicted as classical + classical songs incorrectly predicted as another genre}}
$$

**In general, there is a trade-off to be made between precision and recall**.

At this point, the decision tree is still too complex for interpretability and visualization. Cost-complexicity pruning is now implemented:



```python
path=classifier.cost_complexity_pruning_path(X_train,y_train) #a 2D array that contains alphas and their corresponding node impurities

ccp_alphas=path.ccp_alphas
impurities=path.impurities
```

Note that the last row in the path is the $\alpha$ that prunes the decision tree and leaves it with only one node, so it's always removed:

```python
ccp_alphas=ccp_alphas[:-1]
impurities=impurities[:-1]
```

By setting $\alpha > 0$, cost-complexicity pruning  effectively decreases tree depth and node count (two measures of a decision tree's complexicity). Using a loop, multiple classification trees are created by varying $\alpha$:

```python
classifiers = []
for i in ccp_alphas:
    classifier=DecisionTreeClassifier(random_state=0, ccp_alpha=i)
    classifier.fit(X_train,y_train)
    classifiers.append(classifier)
```

Accuracy is used to evaluate each classifier:

```python
train_scores=[i.score(X_train,y_train) for i in classifiers]
test_scores=[i.score(X_test,y_test) for i in classifiers]
```

```python
fig,axis=plt.subplots(figsize=(7,7))
axis.set_xlabel("Alpha")
axis.set_ylabel("Model Accuracy")
axis.plot(ccp_alphas,train_scores,label="train",drawstyle="steps-post")
axis.plot(ccp_alphas,test_scores,label="test",drawstyle="steps-post")
axis.legend()
```

Test accuracy is highest at around $\alpha=0.002$:

```python
df=pd.DataFrame({"Test Accuracy":test_scores,"Alpha":ccp_alphas})
df["Test Accuracy"].idxmax()
```

Setting $\alpha=0.001605$,

```python
final_classifier=DecisionTreeClassifier(ccp_alpha=0.001605,
                                        random_state=42)
final_model=final_classifier.fit(X_train,y_train)
predictions=final_model.predict(X_test)

print(accuracy_score(y_test,predictions))
```

```python
os.environ["PATH"].split(os.pathsep)
os.environ["PATH"]+=os.pathsep+r"C:\Users\chuat\anaconda3\Library\bin\graphviz"

dot=tree.export_graphviz(final_model,out_file=None)
graph=pydotplus.graph_from_dot_data(dot)  
Image(graph.create_png())
```

Through cost-complexity pruning, the classification tree's accuracy increased from $58\%$ to $64\%$. However, the plot shows that the tree is still too complex. 

There are other parameters that can be optimized, namely stopping criterions such as:

- *max_depth*
- *min_samples_split*
- *min_samples_leaf*

Recall the concept of node purity: each node of the tree must be most of the same class as possible (in principle). Among other stopping criterions, the splits at each node can be specified based on the minimum decrease in impurity. The classification trees above use the default *min_impurity_decrease* of $0$. 

In practice, a series of values for *min_impurity_decrease* are evaluated. Since I want to save hyperparameter optimization for my random forest later, let *min_impurity_decrease* $=0.01$:

```python
final_classifier=DecisionTreeClassifier(ccp_alpha=0.001605,
                                        min_impurity_decrease=0.01,
                                        random_state=42)
final_model=final_classifier.fit(X_train,y_train)
predictions=final_model.predict(X_test)

print(accuracy_score(y_test,predictions))
```

```python
classes=final_model.classes_ #classes
features=data.iloc[:,1:].columns #feature names

dot=tree.export_graphviz(final_model,out_file=None,
                         class_names=classes,feature_names=features,
                         filled=True,rounded=True)
graph=pydotplus.graph_from_dot_data(dot)  
Image(graph.create_png())
```

At the expense of a decrease in accuracy from $64\%$ down to $57\%$, the decision tree becomes interpretable.

#### **Model 2: Bagging (A Special Case of Random Forests)**

```python
baseline=DecisionTreeClassifier()
bagging=BaggingClassifier(base_estimator=baseline,
                          random_state=42)
```

```python
model=bagging.fit(X_train,y_train)
predictions=model.predict(X_test)

print(classification_report(y_test,predictions))
```

#### **Model 3: Random Forest Classification**

```python
numTrees=list(range(1,101)) #n=1,2,...,100
accuracy=[]
for i in numTrees:
  forest=RandomForestClassifier(n_estimators=i,
                                random_state=42)
  model=forest.fit(X_train,y_train)
  predictions=model.predict(X_test)
  accuracy.append(accuracy_score(y_test,predictions))
```

The bagged classification tree above has an accuracy of $69\%$. While trees are still created from bootstrapped training sets, recall from *bagging&RandomForests.ipynb* that random forests differ from bagging by considering only a random subset of the $p$ features at each split of the trees. Indeed, bagging is simply a special case of random forests when all $p$ features are considered at each split.

There are two factors that determine a random forest's accuracy:

- the $n$ number of bootstrapped trees to be created (as with bagging, a large $n$ doesn't lead to overfitting)
- the $m$ number of features considered to be at each split

In a random forest, $1 \leq m < p$. Using loops, the $n$ and $m$ that maximize model accuracy are determined:

```python
numFeatures=list(range(1,9)) #m=1,2,...,8
accuracy1=[]
for i in numFeatures:
  forest=RandomForestClassifier(max_features=i,
                                n_estimators=20,
                                random_state=42)
  model=forest.fit(X_train,y_train)
  predictions=model.predict(X_test)
  accuracy1.append(accuracy_score(y_test,predictions))
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4)) 
fig.suptitle("Accuracy",fontsize=15)
fig.subplots_adjust(top=0.83,wspace=0.4)

ax1.plot(numTrees,accuracy)
ax1.set_title("By # of Bootstrapped Trees")
ax1.set_xlabel("n")

ax2.plot(numFeatures,accuracy1)
ax2.set_title("By # of Features Considered at Each Split")
ax2.set_xlabel("m")
```

Having varied $n$ and $m$ separately, it's important to note that the differences in accuracy are not as substantial as the plots show (see the y axes). In practice, a more methodological approach is to consider $m$ together with several stopping criterions. This is a very memory intensive task depending on how many hyperparameters are optimized at the same time. Let $n=50$.

```python
m=list(range(1,9))
leaves=list(range(5,50,10))
depth=list(range(5,100,5))
splits=list(range(5,50,5))
```

```python
print("Possible Combinations:",len(m)*len(leaves)*len(depth)*len(splits))
```

There are 6840 possible combinations of the $m$ and the three stopping criterions I will be optimizing. Through random search 5-fold cross-validation, $30\%$ of these combinations will be randomly considered:

```python
grid=dict(max_features=m,
          max_depth=depth,
          min_samples_leaf=leaves,
          min_samples_split=splits)
forest=RandomForestClassifier(n_estimators=50,
                              random_state=42)
search=RandomizedSearchCV(estimator=forest,
                          param_distributions=grid,
                          n_iter=int(6840*0.25),
                          cv=3,
                          n_jobs=-1, #all CPUs are used
                          random_state=42)

search.fit(X_train,y_train)
```

```python
search.best_params_
```

```python
forest=RandomForestClassifier(n_estimators=50,
                              max_features=7,
                              max_depth=35,
                              min_samples_leaf=5,
                              min_samples_split=10,
                              random_state=42)
model=forest.fit(X_train,y_train)
predictions=model.predict(X_test)

print(classification_report(y_test,predictions))
```
